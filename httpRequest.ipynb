```python
from bs4 import BeautifulSoup
htmlRUL = 'https://www.ptt.cc/bbs/job/index'
htmlRUL
```

```python
pages = range(1,10)
allurl = list()
urlfirst = 'https://www.ptt.cc/'
for page in pages:
    link = htmlRUL + str(page) + '.html'
    response = requests.get(link)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    a_tags = soup.find_all('a')
    for a in a_tags:
        href = a.get('href')
        if href and href.startswith('/'):
            temp = urlfirst + href
            allurl.append(temp)
```

```python
allurl
```

```python
filtered_urls = [url for url in allurl if url.startswith('https://www.ptt.cc//bbs/job/M.')]
print(filtered_urls)
```

```python
titles = list()
contents = list()
for i in filtered_urls:
    res = requests.get(i)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'html.parser')
    title_tags = soup.find_all(class_='article-meta-value')
    titles.append(title_tags)
    main_content = soup.find_all('div', {'id': 'main-content'})
    content = main_content[0].get_text()
    contents.append(content)
```
